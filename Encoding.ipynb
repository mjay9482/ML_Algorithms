```python
df2 = pd.read_csv('covid_toy.csv')
```

```python
df2.head()
```

this means 10 fever values are missing

```python
df2.isnull().sum()
```

```python
df2['cough'].nunique()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df2.iloc[:,0:5],df2.iloc[:,-1],test_size = 0.2,random_state = 2)
```

```python
X_train
```

##1. Simple Approach

Imputation is performed  to fill missing values in the 'fever' feature.

```python
# Imputation using SimpleImputer for 'fever' feature

si = SimpleImputer()
X_train_fever  = si.fit_transform(X_train[['fever']])

X_test_fever  = si.fit_transform(X_test[['fever']])
X_train_fever.shape
```

Encoding is performed to convert categorical values of 'cough' ('Mild' and 'Strong') into numerical labels.

```python
# Encoding using OrdinalEncoder for 'cough' feature
oe = OrdinalEncoder(categories=[['Mild','Strong']])
X_train_cough = oe.fit_transform(X_train[['cough']])

X_test_cough = oe.fit_transform(X_test[['cough']])

X_train_cough.shape
```

One-hot encoding is performed to convert categorical values of 'gender' and 'city' into binary vectors. The parameter drop='first' is set to drop the first category to avoid multicollinearity.

```python
# One-hot encoding using OneHotEncoder for 'gender' and 'city' features
ohe = OneHotEncoder(drop = 'first',sparse = False)

X_train_gender_city = ohe.fit_transform(X_train[['gender','city']])

X_test_gender_city = ohe.fit_transform(X_test[['gender','city']])

X_train_gender_city.shape
```

```python
# Extracting 'age' feature
X_train_age = X_train.drop(columns = ['gender','fever','cough','city']).values
X_test_age = X_test.drop(columns = ['gender','fever','cough','city']).values

X_train_age.shape
```

```python
X_train_transformed = np.concatenate((X_train_age,X_train_fever,X_train_gender_city,X_train_cough),axis = 1)

X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_test_gender_city,X_test_cough),axis = 1)
X_train_transformed.shape
```

## 2. Using Column Transformer

 ColumnTransformer allows us to apply different preprocessing steps to different columns in dataset, providing a flexible and convenient way to handle data preprocessing.

```python
from sklearn.compose import ColumnTransformer
```

```python
trans = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse = False,drop = 'first'),['gender','city'])
],remainder='passthrough')
```
