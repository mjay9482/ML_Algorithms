```python
# df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
# df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
# df['Year'] = df['Date_of_Journey'].str.split('/').str[2]

df['Date'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
df['Month'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
df['Year'] = df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
df.drop('Date_of_Journey',axis = 1, inplace = True)
```

```python
df.info()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df['Arrival_hour'] = df['Arrival_Time'].apply(lambda x:x.split(':')[0])
df['Arrival_min'] = df['Arrival_Time'].apply(lambda x:x.split(':')[1])

df['Dep_Time'] = df['Dep_Time'].apply(lambda x:x.split(' ')[0])
df['Dep_hour'] = df['Dep_Time'].apply(lambda x:x.split(':')[0])
df['Dep_min'] = df['Dep_Time'].apply(lambda x:x.split(':')[1])
```

```python
df['Arrival_hour'] = df['Arrival_hour'].astype(int)
df['Arrival_min'] = df['Arrival_min'].astype(int)
df.drop('Arrival_Time',axis = 1, inplace = True)

df['Dep_hour'] = df['Dep_hour'].astype(int)
df['Dep_min'] = df['Dep_min'].astype(int)
df.drop('Dep_Time',axis = 1, inplace = True)
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':1, '3 stops':1, '4 stops':1, 'nan': 1})
df.drop('Route', axis = 1, inplace = True)
```

```python
df['Additional_Info'].unique()
```

```python
df[df['Duration'] == '5m']
```

```python
df.drop(6474, axis = 0, inplace = True)
df.drop(2660, axis = 0, inplace = True)
```

```python
df['duration_hour'] = df['Duration'].apply(lambda x:x.split('h')[0])
df['duration_min'] = df['Duration'].apply(lambda x:x.split('h')[1])
df['duration_min'] = df['duration_min'].apply(lambda x:x.split('m')[0])
```

```python
df
```

```python
# Replace empty strings with a default value, e.g., '0'
df['duration_hour'] = df['duration_hour'].replace('', '0')
df['duration_min'] = df['duration_min'].replace('', '0')

df['duration_hour'] = df['duration_hour'].astype(int)
df['duration_min'] = df['duration_min'].astype(int)
df.drop('Duration', axis = 1, inplace = True)
```

```python
df['total_duration']  = df['duration_hour']*60 + df['duration_min']
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
lee = LabelEncoder()
df['Airline'] = lee.fit_transform(df['Airline'])
df['Source'] = lee.fit_transform(df['Source'])
df['Destination'] = lee.fit_transform(df['Destination'])
df['Additional_Info'] = lee.fit_transform(df['Additional_Info'])
```
