```python
import pandas as pd
df = pd.read_csv('breast_cancer_dataset.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,2:]
y = df.iloc[:,1]
```

```python
X = X.dropna(axis = 1)
```

```python
X.head()
```

```python
y.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

## HoldOut Validation Approach

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 0)
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
result = model.score(X_test,y_test)
```

```python
result
```

## KFold Cross Validation

The dataset is divided into k folds, and each fold is used as the validation set once while the remaining k-1 folds are used for training.

```python
from sklearn.model_selection import KFold
model = DecisionTreeClassifier()
kfold_validation = KFold(10)
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
result = cross_val_score(model,X,y,cv = kfold_validation)
result.mean()
```

##Stratified KFold Validation

It ensures that each fold contains approximately the same proportion of class labels as the original dataset.Generally used whenever there is an imbalance of the dataset

```python
from sklearn.model_selection import StratifiedKFold
skfold = StratifiedKFold(n_splits = 5)
model = DecisionTreeClassifier()
scores = cross_val_score(model,X,y,cv = skfold)
np.mean(scores)
```
